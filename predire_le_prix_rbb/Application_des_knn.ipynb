```python
##Importation des librairies 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import torch
import sklearn
```

```python
##ici je vais ouvrir le dataset en question

#ouverture du dataframe
df = pd.read_csv("listings.csv")

df.head(10)
```

```python
#ici on se debarasse en premier des colones inutile
df = df.drop(columns = df.columns[0:5])
df = df.drop(columns = "last_review")
df.head(10)
```

```python
#on suprime les lignes qui possede un nan
df = df.dropna()
df.head(10)
```

```python
df = pd.concat([df, pd.get_dummies(df['room_type'], prefix = 'room_type')], axis=1)
df = pd.concat([df, pd.get_dummies(df['neighbourhood'], prefix = 'neighbourhood')], axis=1)
df = df.drop(columns = ["room_type", 'neighbourhood'])
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing

price = df['price']
df.drop(columns = 'price')
#ppartieion en training et test set 
X = df.to_numpy()
Y = price.to_numpy()

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

score_list = []
for k in range(2, 10):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, Y_train)
    print(knn.score(X_test, Y_test))
    score_list.append(1 - knn.score(X_test, Y_test))
a = [k for k in range(2, 10)]
plt.plot(a, score_list)
plt.show()
```

```python
#test d'une regression lineaire
from sklearn.linear_model import LinearRegression

X = df.to_numpy()
X = preprocessing.scale(X)
Y = price.to_numpy()

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

reg = LinearRegression()
reg.fit(X_train, Y_train)
reg.score(X_test, Y_test)
```
